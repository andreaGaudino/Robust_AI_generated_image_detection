```python
# ========================================
# STEP 2: Environment Check
# ========================================

import sys
import torch

print("=" * 60)
print("🔍 ENVIRONMENT CHECK")
print("=" * 60)

# Python version
print(f"\n✓ Python: {sys.version.split()[0]}")

# PyTorch & MPS
print(f"✓ PyTorch: {torch.__version__}")

if torch.backends.mps.is_available():
    device = torch.device("mps")
    print(f"✓ Device: MPS (M4 Pro GPU)")
    
    # Quick GPU test
    x = torch.randn(1000, 1000, device=device)
    y = torch.matmul(x, x)
    print(f"✓ GPU test: PASSED")
else:
    device = torch.device("cpu")
    print(f"⚠ Device: CPU (slower)")

# Check required libraries
print("\n📚 Libraries:")
libs = [
    'transformers',
    'albumentations', 
    'PIL',
    'sklearn',
    'numpy',
    'pandas',
    'tqdm'
]

all_ok = True
for lib in libs:
    try:
        __import__(lib)
        print(f"  ✓ {lib}")
    except ImportError:
        print(f"  ✗ {lib} - MISSING")
        all_ok = False

print("\n" + "=" * 60)
if all_ok:
    print("✅ ALL READY!")
else:
    print("❌ Install missing libraries:")
    print("   pip install -r requirements.txt")
print("=" * 60)
```
